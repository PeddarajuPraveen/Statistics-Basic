#### Most sales-related variables are highly skewed, with the majority of values concentrated at the lower end and a few extreme outliers pulling the distributions upward. Volume and Average Price indicate that while most products are sold in small quantities at modest prices, a few items have unusually high sales volumes or premium prices. Total Sales Value and Net Sales Value are dominated by a small number of very large transactions, suggesting that a significant share of revenue comes from a few high-value deals. Discount Rate (%) remains low for most cases, though some transactions received exceptionally large discount amounts, likely due to promotions or bulk sales. Overall, the data indicates that regular sales are small and consistent, but business performance is heavily influenced by a limited number of high-value or discounted transactions.

## bar chart for categorical columns

```python
categorical_col=df1.select_dtypes(include=["object"])
```

```python
categorical_col.head()
```

```python
categorical_col.drop("Date", axis=1, inplace=True)
```

```python
for col in categorical_col:
    plt.figure(figsize=(12, 8))
    sns.countplot(data=categorical_col, x=col, order=categorical_col[col].value_counts().index)
    plt.title(f'Distribution of Sales by {col}')
    plt.xlabel(col)
    plt.ylabel('Number of Sales')
    plt.xticks(rotation=50, ha='right')
    plt.tight_layout()
    plt.show()
```

##### Analaysis & Insights
Jeera dominates sales with the highest frequency (90+ sales), so maintain and expand it.

Viva and Gear are in the middle range (~60).

RealU, YouM, Babaji, Vedic all contribute moderately (~45).

Orange & Sumsang lag behind (~30), meaning theyre the weakest performing brands, Focus on promotions to boost their sales.

# conclusion

#### Data preprocessing is crucial because it ensures fair and accurate model training. Standardization puts continuous features on the same scale, preventing dominance by larger-valued variables, while one-hot encoding transforms categorical data into a machine-readable format without implying false order. Together, they improve model performance, interpretability, and reliability.
